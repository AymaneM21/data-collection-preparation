```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_classname(url, selector):
    # Send an HTTP request to the specified URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        
                
        # Use a CSS selector to find the div with the specified class name
        classcontents = soup.select(selector)
        
        results = []

        for classcontent in classcontents:
            # Find and format specific elements in the desired order
            
            # Find the title from the h4 tag
            title = classcontent.find('h4').get_text(strip=True)
            results.append(f"{title} :\n")


            # Find the paragraphs containing the workout schedule
            paragraphs = classcontent.find_all('p')
            for paragraph in paragraphs:
                # Split paragraph into lines based on new lines
                lines = paragraph.get_text("\n", strip=True).splitlines()
                for line in lines:
                    # Check if line starts with a day name
                    if any(line.startswith(day) for day in ["Monday:", "Tuesday:", "Wednesday:", "Thursday:", "Friday:", "Saturday/Sunday:","Saturday","Sunday"]):
                        results.append(f"{line}")
                    else:
                        # Otherwise, add it as a regular paragraph
                        results.append(f"{line}\n")
                results.append("\n")  # Add a blank line after each paragraph



        # Join results with actual newlines for formatting
        formatted_output = "".join(results)
        return formatted_output
    else:
        print(f"Failed to fetch the webpage, status code: {response.status_code}")
        return None

```

PUSH/PULL Exercices :

```python
def extract_workout_routine(url, div_classes):
    """
    Extracts and prints workout routine from specified div classes in the given URL.

    Parameters:
    - url (str): The URL of the webpage to scrape.
    - div_classes (list): List of strings, each string is a class name of the <div> to scrape.

    Returns:
    - None
    """
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    extracted_text = []

    for div_class in div_classes:
        div_content = soup.find('div', class_=div_class)
        if div_content:
            h3_text = div_content.find('h3').text.strip()
            extracted_text.append(h3_text)
            extracted_text.append('')

            main_ul = div_content.find('ul')
            lis = main_ul.find_all('li', recursive=False) if main_ul else []

            for li in lis:
                exercise_name = li.text.split('\n')[0].strip()
                extracted_text.append(exercise_name)

                ul = li.find('ul')
                lis_inner = ul.find_all('li') if ul else []
                for li_inner in lis_inner:
                    li_text = li_inner.text.strip()
                    extracted_text.append(li_text)

                extracted_text.append('')
    return extracted_text

    # Print the extracted text
    for text in extracted_text:
        print(text)
```

```python
# Example usage:
url3 = 'https://www.bodybuildingmealplan.com/chris-bumstead-push-pull-legs/'
div_classes = ['fusion-text fusion-text-10', 'fusion-text fusion-text-11', 
               'fusion-text fusion-text-12', 'fusion-text fusion-text-13', 
               'fusion-text fusion-text-14', 'fusion-text fusion-text-15']

pushpull = extract_workout_routine(url3, div_classes)
pushpulllegs = ""
# Print the extracted text
for text in pushpull:
    pushpulllegs = pushpulllegs + text + "\n"
print(pushpulllegs)
```

5 DAY BRO SPLIT Exercices :

```python
url4 = 'https://www.hevyapp.com/bro-split-workout-program-guide/'


# Fetch the HTML content
response = requests.get(url4)
soup = BeautifulSoup(response.content, 'html.parser')

# Define the selector to target the specific <div> element
selector = 'div.su-note-inner'

# Find all <div> elements with the specified class
div_elements = soup.select(selector)

# Initialize lists to store training days and exercises
training_days = []
exercises = []

# Extract training days and exercises
for div in div_elements:
    # Find all <p> elements within the current <div>
    paragraphs = div.find_all('p')
    
    current_day = None
    current_exercises = []

    for p in paragraphs:
        strong_tag = p.find('strong')
        if strong_tag:
            # If we have a current day and exercises, save them
            if current_day is not None:
                training_days.append(current_day)
                exercises.append(current_exercises)
            
            # Start a new training day
            current_day = strong_tag.text.strip()
            current_exercises = []
        else:
            # Append exercises to the current day's list
            current_exercises.append(p.text.strip())
    
    # Append the last found day and exercises
    if current_day is not None:
        training_days.append(current_day)
        exercises.append(current_exercises)

Brosplit = ""
# Print the results
for i in range(len(training_days)):
    Brosplit = Brosplit + "\n" + training_days[i] + " :" + "\n"
    
    for exercise in exercises[i]:
        Brosplit = Brosplit + "\n" +  exercise + "\n"


print(Brosplit)
```

2000 Calories Meal Plans :

```python
# Replace with the actual URL
url5 = 'https://www.healthline.com/nutrition/2000-calorie-diet#meal-plan'

# Fetch the HTML content
response = requests.get(url5)
soup = BeautifulSoup(response.content, 'html.parser')

# Define the selector to target the specific <div> elements for meal plans
selector = 'div.css-1avyp1d ~ div'

# Find all <div> elements with the specified class that are siblings after css-1avyp1d
div_elements = soup.select(selector)

# Initialize lists to store meal plans and meals
meal_plans = []
meals = []

# Extract meal plans and meals
for div in div_elements:
    # Find all relevant tags within the current <div>
    headers = div.find_all(['h3', 'p', 'ul'])
    
    current_plan = None
    current_meals = []
    plan_count = 0

    for tag in headers:
        if tag.name == 'h3':
            # If we have a current plan and meals, save them
            if current_plan is not None:
                meal_plans.append(current_plan)
                meals.append(current_meals)
            
            # Start a new meal plan
            plan_count += 1
            current_plan = f"Meal Plan {plan_count}"
            current_meals = []
        elif tag.name == 'p':
            # Append meals to the current plan's list
            current_meals.append(tag.text.strip())
        elif tag.name == 'ul':
            items = [li.text.strip() for li in tag.find_all('li')]
            current_meals.append('\n'.join(items))

    # Append the last found plan and meals
    if current_plan is not None:
        meal_plans.append(current_plan)
        meals.append(current_meals)
calories2000 = ""
# Print the results
for i in range(len(meal_plans)):
    calories2000 = calories2000 + "\n" + meal_plans[i] + " :" + "\n" 
    for meal in meals[i]:
        calories2000 = calories2000 + "\n" + meal + "\n"
print(calories2000)

```

# PDF

```python
from fpdf import FPDF
```

```python
# Create a PDF with structured data
class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'Training Programs, Exercises, and Nutrition Plans', 0, 1, 'C')

    def chapter_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, title, 0, 1, 'L')
        self.ln(5)

    def chapter_body(self, body):
        self.set_font('Arial', '', 12)
        self.multi_cell(0, 10, body)
        self.ln()
pdf = PDF()
pdf.add_page()
```

```python
# Define the character replacement dictionary
replacement_dict = {
    '’': "'",
    '–': '-',
    '—': '-',
    '•': '-',
    }
```

```python
def replace_characters(text):
    for char, replacement in replacement_dict.items():
        text = text.replace(char, replacement)
    return text
```

```python
Training_programs = replace_characters(Training_programs)
Exercices = replace_characters(Exercices)
Nutrition = replace_characters(Nutrition)
```

```python
# Add Training Programs
pdf.chapter_title('Training Programs')
pdf.chapter_body(Training_programs)

# Add Exercises
pdf.chapter_title('Exercises')
pdf.chapter_body(Exercices)

# Add Nutrition Plans
pdf.chapter_title('Nutrition Plans')
pdf.chapter_body(Nutrition)


```

```python
# Save the PDF
pdf.output('TrainingPrograms-Exercices-Meals.pdf')

print("PDF generated successfully.")
```
